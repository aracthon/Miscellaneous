```python
from scipy.stats import bernoulli
from scipy.stats import binom
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

# ----------- part a) of problems ----------- #
n, p = 5, 0.5
mean, var, skew,  kurt = binom.stats(n, p, moments='mvsk')

x = [0, 1, 2, 3, 4, 5]

plt.figure(figsize=(3,3))
plt.subplot(1,1,1)
plt.plot(x, binom.pmf(x, n, p), 'o', ms=5)
plt.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5)
plt.title('Binomial PMF')
plt.xlabel('Value of k')
plt.ylabel('probabilities of P(X_5 = k)')
plt.xlim(-1, 6)
plt.show()
```

```python
# ----------- part b) of problems ----------- #
from collections import defaultdict
SampleNum = 100
SampleSpace = np.random.binomial(n, p, SampleNum)
SampleNumDict = defaultdict(float)

for i in SampleSpace:
    SampleNumDict[i] += 1

for k, v in SampleNumDict.items():
    SampleNumDict[k] = v * 1.0 / SampleNum

plt.figure(figsize=(3,3))
plt.subplot(1, 1, 1)
plt.title('Samples PMF with J = 100')
plt.xlabel('Value of k')
plt.ylabel('frequencies of Samples')
plt.bar(range(len(SampleNumDict)), SampleNumDict.values(), align='center', width=0.1)
plt.show()
```
